```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

inputFile = "bitsize.csv" 
data = pd.read_csv(inputFile) 
```

```python
x = ['6', '6.125',  '6.75', '7.875', '8.5', '8.75']
y = []

count = 0
index = 0
for num in x :
    for i in data.bitsize :
        if (i == float(num)) :
            count = count + 1
    y.insert(index, count)
    count = 0
    index = index + 1
print(y)
```

```python
fig = plt.figure(figsize = (10, 5))
 
plt.bar(x, y, color ='dodgerblue',
        width = 0.4)

plt.title("Bit Size of Rigs During Backouts")
plt.xlabel("LAT Bit Size")
plt.ylabel("Number of Backouts")


plt.show()
```

```python
fig, ax = plt.subplots()

#(counts, percentage, colors, pctdistance, labeldistance)
ax.pie(y, labels=x, autopct='%1.1f%%', colors=['lightyellow', 'peachpuff', 'lightskyblue', 'lavender', 'mistyrose', 'honeydew'])

plt.show()
```
