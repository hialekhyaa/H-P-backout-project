```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

inputFile = "size.csv" 
data = pd.read_csv(inputFile) 
```

```python
x = ['6.75', '7.875', '8.5', '8.75']
y = []

count = 0
index = 0
for num in x :
    for i in data.bitsize :
        if (i == float(num)) :
            count = count + 1
    y.insert(index, count)
    count = 0
    index = index + 1
print(y)
```

```python
#size of figure
fig = plt.figure(figsize = (10, 5))
 
#creates bar chart
plt.bar(x, y, color ='dodgerblue',
        width = 0.4)
 
#labels
plt.title("Bit Size of Rigs During Backouts")
plt.xlabel("Bit Size")
plt.ylabel("Number of Backouts")

#display
plt.show()
```

```python
fig, ax = plt.subplots()

#(counts, percentage, colors, pctdistance, labeldistance)
ax.pie(y, labels=x, autopct='%1.1f%%', colors=['lightskyblue', 'peachpuff', 'mistyrose', 'lavender'])

plt.show()
```
