```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

inputFile = "size.csv" 
data = pd.read_csv(inputFile) 
```

```python
x = ['4.5', '5', '5.5']
y = []

count = 0
index = 0
for num in x :
    for i in data.dpsize :
        if (i == float(num)) :
            count = count + 1
    y.insert(index, count)
    count = 0
    index = index + 1
print(y)
```

```python
fig = plt.figure(figsize = (10, 5))
 

plt.bar(x, y, color ='mediumvioletred',
        width = 0.4)
 
plt.title("Drill Pipe Size of Rigs During Backouts")
plt.xlabel("Pipe Size")
plt.ylabel("Number of Backouts")



plt.show()
```

```python

fig, ax = plt.subplots()
ax.pie(y, labels=x, autopct='%1.1f%%', colors=['paleturquoise', 'palevioletred', 'lemonchiffon'])

plt.show()
```
